Seasonal Trend to detect patterns in bird sightings across different seasons or years.

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
```

```python
dfg['Season'] = dfg['Date'].dt.month.apply(get_season)
```

```python
seasonal_counts = dfg.groupby(['Year', 'Season']).size().unstack(fill_value=0)
ax=seasonal_counts.plot(kind='bar', stacked=True, colormap = 'coolwarm', figsize=(10, 6))
plt.title('Seasonal Trends of Bird Sightings- Grassland')
plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.xticks(rotation=0)
plt.legend(title='Season')
plt.tight_layout()
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="center", fontsize=10,fontweight='bold', padding=3)
plt.show()
```

```python
dff['Season'] = dff['Date'].dt.month.apply(get_season)
seasonal_counts = dff.groupby(['Year', 'Season']).size().unstack(fill_value=0)
ax=seasonal_counts.plot(kind='bar', stacked=True, colormap = 'coolwarm', figsize=(10, 6))
plt.title('Seasonal Trends of Bird Sightings- Forest')
plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.xticks(rotation=0)
plt.legend(title='Season')
plt.tight_layout()
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="center", fontsize=10,fontweight='bold', padding=3)
plt.show()
```


Bird Sightings by Time

```python
dfg['Start_Hour'] = dfg['Start_Time'].dt.hour
# Calculate the duration of the observation in minutes
dfg['Duration'] = (dfg['End_Time'] - dfg['Start_Time']).dt.total_seconds() / 60
# Define time categories based on Start_Hour
def get_time_of_day(hour):
    if 5 <= hour < 9:
        return 'Morning'
    elif 9 <= hour < 12:
        return 'Late Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 20:
        return 'Evening'
    else:
        return 'Night'

# Apply the function to classify time of day
dfg['Time_of_Day'] = dfg['Start_Hour'].apply(get_time_of_day)

# Group by Time_of_Day to count the number of bird sightings
time_of_day_counts = dfg.groupby('Time_of_Day').size()

# Plot the results
ax=time_of_day_counts.plot(kind='bar', color='skyblue', figsize=(10, 6))

# Customize the plot
plt.title('Bird Sightings by Time of Day - Grassland')
plt.xlabel('Time of Day')
plt.ylabel('Number of Sightings')
plt.xticks(rotation=45)
plt.tight_layout()
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="center", fontweight='bold', fontsize=10, padding=3)
plt.show()

# Optionally, to see the average duration for each time of day
avg_duration_by_time = dfg.groupby('Time_of_Day')['Duration'].mean()
print(avg_duration_by_time)
```

Importing ttest_ind, to check if there is a statistically significant difference in observation distance between two groups

Impact of Disturbance on Bird Observations

```python
from scipy.stats import ttest_ind
disturbance_counts = dff["Disturbance"].value_counts().reset_index()
disturbance_counts.columns = ["Disturbance", "Bird_Count"]

# Bar plot: Bird observations under different disturbance levels
plt.figure(figsize=(6, 4))
ax=sns.barplot(data=disturbance_counts,  x="Disturbance", y="Bird_Count", palette="coolwarm")
plt.xlabel("Disturbance Level", fontsize=12, fontweight="bold")
plt.ylabel("Number of Bird Observations", fontsize=12, fontweight="bold")
plt.title("Impact of Disturbance on Bird Observations_forest", fontsize=14, fontweight="bold")
plt.xticks(rotation=45)
for container in ax.containers:
    ax.bar_label(container, fmt="%.0f", label_type="center", fontweight='bold', fontsize=10, padding=3)
plt.show()

distance_mapping = {"<= 50 Meters": 25, "50 - 100 Meters": 75}
dff["Distance_Numeric"] = dff["Distance"].map(distance_mapping)
low_disturbance = dff[dff["Disturbance"] == "No Effect"]["Distance_Numeric"]
high_disturbance = dff[dff["Disturbance"] == "Slight effect on count"]["Distance_Numeric"]
t_stat, p_value = ttest_ind(low_disturbance, high_disturbance, nan_policy="omit")

print(f"T-test result: t={t_stat:.2f}, p={p_value:.4f}")
if p_value < 0.05:
    print("There is a statistically significant difference in bird observation distances based on disturbance level.")
else:
    print("No significant difference in observation distances due to disturbance.")
```
